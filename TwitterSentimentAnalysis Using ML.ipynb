Stemming

Stemming is the process of reducing a word to its Root word

```python
port_stem = PorterStemmer()
```

```python
def stemming(content):

  stemmed_content = re.sub('[^a-zA-Z]',' ', content)
  stemmed_content = stemmed_content.lower()
  stemmed_content = stemmed_content.split()
  stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english') ]
  stemmed_content = ' '.join(stemmed_content)

  return stemmed_content
```

```python
twitter_data['stemmed_content'] = twitter_data['text'].apply(stemming)  #50 mins
```

```python
print(twitter_data['stemmed_content'])
```

```python
print(twitter_data['target'])
```

```python
# seperating the data and label

X = twitter_data['stemmed_content'].values
Y = twitter_data['target'].values
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data to training data test data

```python
X_train , X_test , Y_train ,  Y_test = train_test_split(X, Y , test_size=0.2 , stratify=Y , random_state = 2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(X_train)
print(X_test)
```

```python
# converting the textual data to numerical data

vectorizer = TfidfVectorizer()

# Assuming X_train and X_test are your textual data
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

Training the MachineLearning Model

Logistic Regression

```python
model = LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train,Y_train)
```

Model Evaluation

Accuracy score

```python
#accuracy score on the training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train , X_train_prediction)
```

```python
print('Accuracy score on the training data :' , training_data_accuracy)
```

```python
#accuracy score on the test data

X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test , X_test_prediction)
```

```python
print('Accuracy score on the test data:', test_data_accuracy)
```

Model accuracy = 77.8%

```python

def predict_tweets():
  num_tweets = int(input("Enter the number of tweets to analyze: "))
  for i in range(num_tweets):
    tweet = input(f"Enter tweet {i+1}: ")
    # Preprocess the tweet
    stemmed_tweet = stemming(tweet)
    # Transform the tweet using the trained vectorizer
    vectorized_tweet = vectorizer.transform([stemmed_tweet])
    # Predict the sentiment
    prediction = loaded_model.predict(vectorized_tweet)
    print(prediction[0])

predict_tweets()
```
